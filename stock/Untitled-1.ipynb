```python
%pip install pandas
import pandas as pd
import os

# Define the directory path
folder_path = r'C:\Users\Haritha\OneDrive\Desktop\guvi\stock'
output_file = os.path.join(folder_path, 'combined_stock_data.csv')

all_data = []

# Loop through files in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    
    try:
        # Load Excel files
        if filename.endswith('.xlsx') or filename.endswith('.xls'):
            df = pd.read_excel(file_path)
            all_data.append(df)
        
        # Load existing CSVs (to merge them)
        elif filename.endswith('.csv') and filename != 'combined_stock_data.csv':
            df = pd.read_csv(file_path)
            all_data.append(df)
            
    except Exception as e:
        print(f"Could not read {filename}: {e}")

# Combine and save
if all_data:
    combined_df = pd.concat(all_data, ignore_index=True)
    combined_df.to_csv(output_file, index=False)
    print(f"Success! Saved to: {output_file}")
else:
    print("No compatible files found in the folder.")
```

```python
import pandas as pd
import os
import yaml

folder_path = r'C:\Users\Haritha\OneDrive\Desktop\guvi\stock'
output_file = os.path.join(folder_path, 'combined_stock_data.csv')

all_data = []

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    
    # Handle Excel
    if filename.endswith(('.xlsx', '.xls')):
        all_data.append(pd.read_excel(file_path))
    
    # Handle YAML
    elif filename.endswith(('.yaml', '.yml')):
        with open(file_path, 'r') as f:
            data = yaml.safe_load(f)
            # Converts the YAML dictionary/list to a DataFrame
            all_data.append(pd.DataFrame(data))

if all_data:
    pd.concat(all_data, ignore_index=True).to_csv(output_file, index=False)
    print(f"Done! Created: {output_file}")
else:
    print("Still no matching files. Are the files definitely inside the 'stock' folder?")
```

```python
import pandas as pd
import os

# Define the paths
input_file = r'C:\Users\Haritha\OneDrive\Desktop\guvi\stock\combined_stock_data.csv'
output_dir = r'C:\Users\Haritha\OneDrive\Desktop\guvi\stock\tickers_split'

# Create a new folder for the split files so they don't clutter your main folder
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Load the combined data
try:
    df = pd.read_csv(input_file)
    
    # Identify the column name for tickers (adjust 'Ticker' if your column name is different)
    ticker_column = 'Ticker' 
    
    if ticker_column not in df.columns:
        print(f"Error: Could not find '{ticker_column}' column. Available columns: {list(df.columns)}")
    else:
        # Group by ticker and save each group
        for ticker, group in df.groupby(ticker_column):
            # Clean ticker name for filename (remove special characters if any)
            clean_ticker = str(ticker).strip().replace("/", "_")
            file_name = f"{clean_ticker}.csv"
            file_path = os.path.join(output_dir, file_name)
            
            group.to_csv(file_path, index=False)
            print(f"Created file for: {clean_ticker}")

        print(f"\nSuccess! All files are saved in: {output_dir}")

except FileNotFoundError:
    print("The combined_stock_data.csv file was not found. Please run the previous merge script first.")
except Exception as e:
    print(f"An error occurred: {e}")
```
